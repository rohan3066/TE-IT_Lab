```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
df.columns
```

```python
x=df.iloc[:,3:]
x
```

```python
x
```

```python
plt.title('unclustered data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
```

```python
km=KMeans(n_clusters=6)
```

```python
x.shape
```

```python
km.fit_predict(x)
```

```python
km.inertia_
```

```python
sse=[]
for k in range(1,16):
    km=KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('Elbow Method')
plt.xlabel('Value of K')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker='.',color='red')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silh=[]
for k in range(2,16):
    km=KMeans(n_clusters=k)
    labels=km.fit_predict(x)
    score=silhouette_score(x,labels)
    silh.append(score)
```

```python
silh
```

```python
plt.title('Silhoutte Method')
plt.xlabel('Value of K')
plt.ylabel('Silhoutte score')
plt.grid()
plt.xticks(range(2,16))
plt.bar(range(2,16),silh,color='red')
```

```python
km=KMeans(n_clusters=5,random_state=0)
```

```python
labels=km.fit_predict(x)
```

```python
labels

cent=km.cluster_centers_
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('unclustered data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])


plt.subplot(1,2,2)
plt.title('clustered data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=labels)
plt.scatter(cent[:,0],cent[:,1],s=100,color='k')
```

```python
cent
```

```python
agl=AgglomerativeClustering(n_clusters=5)
```

```python
alabels=agl.fit_predict(x)
```

```python
alabels
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('AgglomerativeClustering data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=alabels)
plt.scatter(cent[:,0],cent[:,1],s=100,color='red')


plt.subplot(1,2,2)
plt.title('Kmeans data')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=labels)
plt.scatter(cent[:,0],cent[:,1],s=100,color='k')
```
